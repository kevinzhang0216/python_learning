```python
# 加入库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods = 1000))
```

```python
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
# 定义名称  x轴 y轴 标签
ts.plot(title = 'qushitu')
plt.ylabel('length')
plt.xlabel('time')
```

```python
# 画一个dataFrame的图
df = pd.DataFrame(np.random.randn(1000,4), index = ts.index, columns= list('ABCD'))
```

```python
df = df.cumsum()
df.plot()
```

```python
# 默认使用索引作为x 其他作为y
# 制定x y
df3 = pd.DataFrame(np.random.randn(1000,2),columns=list('BC')).cumsum()
df3.plot(x='B', y = 'C',title = 'test')
plt.ylabel('length')
plt.xlabel('time')
```

# 散点图
1. 要定义x ，y

```python
df = pd.DataFrame(np.random.rand(50,4),columns=list('abcd'))
```

```python
df.plot.scatter(x = 'a',y='d')
```

```python
# 两组数据化为一个数据 ，在一个图中做两组数据
ax = df.plot.scatter(x='a',y='b',color='DarkBlue', label = 'Group 1')
df.plot.scatter(x='c', y='d', color= 'DarkGreen', label = 'Group 2', ax =ax)
```

```python
# 对散点图大小进行设置
df.plot.scatter(x='a',y='b', s=df['c']*200)
```
