# 数据合并
1. 数据合并有两个方式
2. 一个是列的合并
3. 一个是行的合并

```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.DataFrame({ 'one': ['a','b','a','a','c','b'],
                      'two':range(6)
    })
```

```python
data2 = pd.DataFrame({ 'one': ['a','b','c','d'],
                      'two':range(10,14)
    })
```

```python
data1
```

```python
data2
```

```python
data = pd.merge(data1, data2, on = 'one')
# 根据one这一列进行合并对齐 两方都有才保存下来
```

```python
data
```

```python
# 制定使用哪列来合并 根据相同一列来合并
data1 = pd.DataFrame({ 'one1': ['a','b','a','a','c','b'],
                      'two1':range(6)
    })
data2 = pd.DataFrame({ 'one2': ['a','b','c','d'],
                      'two2':range(10,14)
    })
data = pd.merge(data1, data2, left_on = 'one1', right_on = 'one2')
```

```python
data
```

```python
# 只有一列名称一致

data1 = pd.DataFrame({ 'one': ['a','b','a','a','c','b'],
                      'two1':range(6)
    })
data2 = pd.DataFrame({ 'one': ['a','b','c','d'],
                      'two2':range(10,14)
    })
data = pd.merge(data1, data2)
```

```python
data
```

```python
# 所有的数据都要保存下来 ,所有的数据都要保存下来

data1 = pd.DataFrame({ 'one': ['a','b','a','a','c','b'],
                      'two1':range(6)
    })
data2 = pd.DataFrame({ 'one': ['a','b','c','d'],
                      'two2':range(10,14)
    })
data = pd.merge(data1, data2, how = 'outer')
```

```python
data
```

```python
# 根据 右边的数据进行合并

data1 = pd.DataFrame({ 'one': ['a','b','a','a','c','e'],
                      'two1':range(6)
    })
data2 = pd.DataFrame({ 'one': ['a','b','c','d'],
                      'two2':range(10,14)
    })
data = pd.merge(data1, data2, how = 'right')
```

```python
data
```

```python
# 根据 右边的数据进行合并

data1 = pd.DataFrame({ 'one': ['a','b','a','a','c','e'],
                      'two1':range(6)
    })
data2 = pd.DataFrame({ 'one': ['a','b','c','d'],
                      'two2':range(10,14)
    })
data = pd.merge(data1, data2, how = 'left')
```

```python
data
```

## 纵向合并
1. concat 使用的是列表 [data1, data2]

```python
data1 = pd.DataFrame(np.random.randn(3,4), columns=['a','b','c','d'])
data2 = pd.DataFrame(np.random.randn(2,3), columns=['b','d','a'])
```

```python
data1
```

```python
data2
```

```python
data = pd.concat([data1, data2], sort = True)
```

```python
data
```

```python
# 重新排序
data = pd.concat([data1, data2], sort = True, ignore_index= True)
```

```python
data
```

```python
data.append(pd.Series([1,2,3],index= ['a','b','d']),ignore_index=True)
```
