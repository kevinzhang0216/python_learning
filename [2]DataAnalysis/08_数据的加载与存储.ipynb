### 标题的问题
1. 分格符号
2. 没有标题， header =  none
3. 制定columns ，用names = ['one','two']
4. 制定哪一行为 表头
5. 存储 index =False

```python
# 没有columns
data = pd.read_csv('test1.csv', header =None)
```

```python
data.head()
```

```python
# 制定columns
data = pd.read_csv('test1.csv', header =None, names = ['one','two','three','four','five','six'])
```

```python
data.head()
```

```python
# 制定哪一类是索引
data = pd.read_csv('test1.csv',header=None)
```

```python
data.head()
```

```python
# 指定那一列为 index
data = pd.read_csv('test1.csv', index_col = 1)
```

```python
data.head()
```

```python
# 跳过某一部分 跳过某行标题
data = pd.read_csv('test1.csv', skiprows=[0,5], header=None)
```

```python
data.head()
```

```python
# 读某三行的数据
data = pd.read_csv('test1.csv', nrows = 4)
```

```python
data
```
