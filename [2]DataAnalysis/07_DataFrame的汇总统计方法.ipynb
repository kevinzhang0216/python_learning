## DataFrame使用函数


```python
import pandas as pd
import numpy as np 
```

```python
data = pd.DataFrame(np.random.randn(9,6),columns=list('abcdef'))
```

```python
data
```

```python
# 应用一个函数
data.apply(np.cumsum)                        # 按照列执行
# cumsum是累加和 对每一类进行运算
```

```python
data.apply(lambda x : x.max() - x.min())  # 都是应用在列上
```

```python
#在行上使用 axis =1 是列与列之间进行比较运算，因此得到的是每一行的最大值减去最小值
data.apply(lambda x : x.max()-x.min(), axis = 1)
```

```python
data['a'].apply(lambda x : x +10) # 对每一列进行运算
#data[‘a’]是一个series
```

```python
data
```

```python
# 对每一个元素 进行操作  applymap
data.applymap(lambda x : x+100)
```

```python
data
```

```python
# 求和 sum,默认是求每一列的和
data.sum()
```

```python
data.sum(1) # 求一行的和
```

```python
# 最大值得索引 默认是选择每一列的最大值
data.idxmax(1) 
```

```python
data['a'][data.a.idxmax()]
```

```python
data
```

## 找到唯一值

```python
data = pd.DataFrame(np.random.randint(5,10,(5,6)), columns=list('abcdef'))
```

```python
data
```

```python
# a列中有哪些唯一值
data.a.unique()
```

```python
#得到每一列中，数值的重复次数
data.a.value_counts()
```

```python
# 这也是个顶级方法
pd.value_counts(data.b)
```

```python
# isin 函数 判断是否在集合中
data.a[data.a.isin([17,18,19])]
```

```python
data.a
```

```python
data.cumsum()
```

```python
data
```

```python
data.sum()
```

## 层次化索引

```python
data =  pd.DataFrame(np.random.randn(6,4), index = [['m','m','m','m','n','n'],list('abcdef')])
```

```python
data
```

```python
# unsatack 方法
data.unstack()
```

```python
data.unstack(0)
```

```python
data.unstack(0).stack()
```
